#  Using FABRIC Persistent Storage<br>
<br>
Your compute nodes can include persistent Project specific storage volumes.<br>
These volumes can be created on request at various sites through<br>
the (Help Portal)[https://fabric-testbed.atlassian.net/servicedesk/customer/portal/2/group/12].<br>
<br>
This example notebook will demonstrate how to attach a persistent storage to a compute node in your slice.<br>


## Setup the Experiment<br>
<br>
#### Import FABRIC API

In[1]:

```python
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
```

```python
fablib = fablib_manager()
```

```python
fablib.show_config()
```

## Create a Node<br>
<br>
The cell below creates a slice that contains a single node with a 1TB NVMe device.<br>
<br>
<br>
### Set the Slice Name and FABRIC Site

In[ ]:

```python
from fabrictestbed.slice_editor import ComponentModelType
```

```python
slice_name = "MySlice"
node_name = 'Node1'
storage_name = 'vol1'
```

In[ ]:

```python
try:
    # Create Slice
    slice = fablib.new_slice(name=slice_name)

    # Add node
    node = slice.add_node(name=node_name, site="RENC")

    # Add an NVME Drive
    node.add_storage(name=storage_name)

    # Submit Slice Request
    slice.submit()
except Exception as e:
    print(f"Exception: {e}")
```

## Get the Slice

In[ ]:

```python
try:
    slice = fablib.get_slice(name=slice_name)
    print(f"{slice}")
except Exception as e:
    print(f"Exception: {e}")
```

## Get the Node<br>
<br>
Retrieve the storage device name and mount it<br>


In[ ]:

```python
try:
    node = slice.get_node(node_name)
    print(f"{node}")
    storage = node.get_storage(storage_name)
    print(f"{storage}")
    print(f"{storage.get_device_name()}")
    stdout, stderr = node.execute(f"sudo mkdir -p /mnt/{storage_name};"
                                  f"sudo mount {storage.get_device_name()} /mnt/{storage_name}")
    print(stdout)
    print(stderr)
```

```python
except Exception as e:
    print(f"Exception: {e}")
```

## Cleanup Your Experiment

In[ ]:

```python
try:
    slice = fablib.get_slice(name=slice_name)
    slice.delete()
except Exception as e:
    print(f"Exception: {e}")
```
