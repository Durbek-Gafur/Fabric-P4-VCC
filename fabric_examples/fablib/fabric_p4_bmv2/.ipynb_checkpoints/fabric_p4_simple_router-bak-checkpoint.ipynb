### Utility Functions

```python
import paramiko
from ipaddress import ip_address, IPv4Address

bastion_public_addr = 'bastion-1.fabric-testbed.net'
bastion_private_ipv4_addr = '192.168.11.226'
bastion_private_ipv6_addr = '2600:2701:5000:a902::c'

def validIPAddress(IP: str) -> str:
    try:
        return "IPv4" if type(ip_address(IP)) is IPv4Address else "IPv6"
    except ValueError:
        return "Invalid"

def execute_script(node_username, node, script):
    import paramiko

    try:
        management_ip = str(node.get_property(pname='management_ip'))
        #print("Node {0} IP {1}".format(node.name, management_ip))

        key = paramiko.RSAKey.from_private_key_file(ssh_key_file_priv)

        bastion=paramiko.SSHClient()
        bastion.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        bastion.connect(bastion_public_addr, username=bastion_username, key_filename=bastion_key_filename)

        bastion_transport = bastion.get_transport()
        if validIPAddress(management_ip) == 'IPv4':
            src_addr = (bastion_private_ipv4_addr, 22)
        elif validIPAddress(management_ip) == 'IPv6':
            src_addr = (bastion_private_ipv6_addr, 22)
        else:
            return 'Management IP Invalid: {}'.format(management_ip)

        dest_addr = (management_ip, 22)
        bastion_channel = bastion_transport.open_channel("direct-tcpip", dest_addr, src_addr)


        client = paramiko.SSHClient()
        client.load_system_host_keys()
        client.set_missing_host_key_policy(paramiko.MissingHostKeyPolicy())
        client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

        client.connect(management_ip,username=node_username,pkey = key, sock=bastion_channel)

        stdin, stdout, stderr = client.exec_command('echo \"' + script + '\" > script.sh; chmod +x script.sh; sudo ./script.sh')
        stdout_str = str(stdout.read(),'utf-8').replace('\\n','\n')
        #print ('')
        #print (str(stdout.read(),'utf-8').replace('\\n','\n'))
        #print (stdout_str)

        client.close()
    except Exception as e:
        print(str(e))
        return str(e)

    return stdout_str


def upload_file(node_username, node, local_file_path, remote_file_path):
    import paramiko

    try:
        management_ip = str(node.get_property(pname='management_ip'))
        #print("Node {0} IP {1}".format(node.name, management_ip))

        key = paramiko.RSAKey.from_private_key_file(ssh_key_file_priv)

        bastion=paramiko.SSHClient()
        bastion.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        bastion.connect(bastion_public_addr, username=bastion_username, key_filename=bastion_key_filename)

        bastion_transport = bastion.get_transport()
        if validIPAddress(management_ip) == 'IPv4':
            src_addr = (bastion_private_ipv4_addr, 22)
        elif validIPAddress(management_ip) == 'IPv6':
            src_addr = (bastion_private_ipv6_addr, 22)
        else:
            return 'Management IP Invalid: {}'.format(management_ip)

        dest_addr = (management_ip, 22)
        bastion_channel = bastion_transport.open_channel("direct-tcpip", dest_addr, src_addr)


        client = paramiko.SSHClient()
        client.load_system_host_keys()
        client.set_missing_host_key_policy(paramiko.MissingHostKeyPolicy())
        client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

        client.connect(management_ip,username=node_username,pkey = key, sock=bastion_channel)

        ftp_client=client.open_sftp()
        file_attributes = ftp_client.put(local_file_path, remote_file_path)
        ftp_client.close()

        client.close()
    except Exception as e:
        print(str(e))
        return str(e)

    return file_attributes

def download_file(node_username, node, local_file_path, remote_file_path):
    import paramiko

    try:
        management_ip = str(node.get_property(pname='management_ip'))
        #print("Node {0} IP {1}".format(node.name, management_ip))

        key = paramiko.RSAKey.from_private_key_file(ssh_key_file_priv)

        bastion=paramiko.SSHClient()
        bastion.set_missing_host_key_policy(paramiko.AutoAddPolicy())
        bastion.connect(bastion_public_addr, username=bastion_username, key_filename=bastion_key_filename)

        bastion_transport = bastion.get_transport()
        if validIPAddress(management_ip) == 'IPv4':
            src_addr = (bastion_private_ipv4_addr, 22)
        elif validIPAddress(management_ip) == 'IPv6':
            src_addr = (bastion_private_ipv6_addr, 22)
        else:
            return 'Management IP Invalid: {}'.format(management_ip)

        dest_addr = (management_ip, 22)
        bastion_channel = bastion_transport.open_channel("direct-tcpip", dest_addr, src_addr)


        client = paramiko.SSHClient()
        client.load_system_host_keys()
        client.set_missing_host_key_policy(paramiko.MissingHostKeyPolicy())
        client.set_missing_host_key_policy(paramiko.AutoAddPolicy())

        client.connect(management_ip,username=node_username,pkey = key, sock=bastion_channel)

        ftp_client=client.open_sftp()
        file_attributes = ftp_client.get(local_file_path, remote_file_path)
        ftp_client.close()

        client.close()
    except Exception as e:
        print(str(e))
        return str(e)

    return file_attributes

import time
def wait_for_slice(slice,timeout=180,interval=10,progress=False):
    timeout_start = time.time()

    if progress: print("Waiting for slice .", end = '')
    while time.time() < timeout_start + timeout:
        return_status, slices = slice_manager.slices(excludes=[SliceState.Dead,SliceState.Closing])

        if return_status == Status.OK:
            slice = list(filter(lambda x: x.slice_name == slice_name, slices))[0]
            if slice.slice_state == "StableOK":
                if progress: print(" Slice state: {}".format(slice.slice_state))
                return slice
            if slice.slice_state == "Closing" or slice.slice_state == "Dead":
                if progress: print(" Slice state: {}".format(slice.slice_state))
                return slice    
        else:
            print(f"Failure: {slices}")
        
        if progress: print(".", end = '')
        time.sleep(interval)
    
    if time.time() >= timeout_start + timeout:
        if progress: print(" Timeout exceeded ({} sec). Slice: {} ({})".format(timeout,slice.slice_name,slice.slice_state))
        return slice   
    
    
def ping_test(node, ping_target_ip, verbose=False):
    if verbose: print("Ping Test: {}".format(ping_target_ip),end='')

    #rtt min/avg/max/mdev = 0.063/0.119/0.189/0.053 ms
    #output = "Information about latency with ping: \n"
    
    try:
        #ping up
        raw_output = execute_script(username, node, 'ping -c 3 ' + ping_target_ip + ' | grep rtt')
        #raw_output = stdout.read().decode("utf-8")
        raw_data = raw_output.split(" ")[3]
        data_array = raw_data.split("/")
        avg_rtt = data_array[1]

        if verbose: print(", ping success! avg_rtt: {}".format(avg_rtt))
    except Exception as e:
        if verbose: print(", ping failed: {}".format(str(e)))
        return float(0)
        
    return float(avg_rtt)
  
```

## Configure Slice Parameters



```python
# Slice 
slice_name = 'P4_SimpleRouter'

# Router
router_name = "router"
router_cores = 4
router_ram = 16
router_disk = 50

# Nodes
node_count = 4
node_cores = 2
node_ram = 8
node_disk = 10

# All node properties
username = 'ubuntu'
image = 'default_ubuntu_20'
image_type = 'qcow2'
site = 'MAX'

```

```python
from fabrictestbed.slice_editor import ExperimentTopology, Capacities, ComponentType, ComponentModelType, ServiceType
# Create topology
experiment = ExperimentTopology()

# Add Router
router = experiment.add_node(name=router_name, site=site)
cap = Capacities()
cap.set_fields(core=router_cores, ram=router_ram, disk=router_disk)
router.set_properties(capacities=cap, image_type=image_type, image_ref=image)

router_interfaces = {}
router.add_component(model_type=ComponentModelType.SmartNIC_ConnectX_5, name="nic1")
router.add_component(model_type=ComponentModelType.SmartNIC_ConnectX_5, name="nic2")
router_interfaces['iface0'] = list(filter(lambda x : x.name == "nic1-p1", router.interface_list))[0]
router_interfaces['iface1'] = list(filter(lambda x : x.name == "nic1-p2", router.interface_list))[0]
router_interfaces['iface2'] = list(filter(lambda x : x.name == "nic2-p1", router.interface_list))[0]
router_interfaces['iface3'] = list(filter(lambda x : x.name == "nic2-p2", router.interface_list))[0]

# Add nodes
for node_num in range(0, node_count):
    print("Adding node {}".format(node_num))
    node_name = "Node" + str(node_num)
    nic_name = "nic" + str(node_num)
    net_name = "net" + str(node_num)
    router_iface_name = "iface" + str(node_num)
    
    node = experiment.add_node(name=node_name, site=site)
    
    cap = Capacities()
    cap.set_fields(core=node_cores, ram=node_ram, disk=node_disk)
    node.set_properties(capacities=cap, image_type=image_type, image_ref=image)

    # Add nets
    node.add_component(model_type=ComponentModelType.SharedNIC_ConnectX_6, name=nic_name)
    node_interface = list(filter(lambda x : x.name == nic_name+"-p1", node.interface_list))[0]
    
    #router.add_component(model_type=ComponentModelType.SharedNIC_ConnectX_6, name=nic_name)
    route_interface = router_interfaces[router_iface_name]

    #print("node_interface: {}".format(node_interface))
    #print("route_interface: {}".format(route_interface))
          
    experiment.add_network_service(name=net_name, nstype=ServiceType.L2Bridge, interfaces=[route_interface, node_interface] )
    #print("router interface_list: {}".format(list(router.interface_list)))
    #print("router interface_net: {}".format(route_net))

# Generate Slice Graph
slice_graph = experiment.serialize()

# Request slice from Orchestrator
return_status, slice_reservations = slice_manager.create(slice_name=slice_name, 
                                            slice_graph=slice_graph, 
                                            ssh_key=ssh_key_pub)

if return_status == Status.OK:
    slice_id = slice_reservations[0].get_slice_id()
    print("Submitted slice creation request. Slice ID: {}".format(slice_id))
else:
    print(f"Failure: {slice_reservations}")
  
```

## Get the Slice

```python
return_status, slices = slice_manager.slices(excludes=[SliceState.Dead,SliceState.Closing])

if return_status == Status.OK:
    slice = list(filter(lambda x: x.slice_name == slice_name, slices))[0]
    slice = wait_for_slice(slice, timeout=300, progress=True)

print()
print("Slice Name : {}".format(slice.slice_name))
print("ID         : {}".format(slice.slice_id))
print("State      : {}".format(slice.slice_state))
print("Lease End  : {}".format(slice.lease_end))
```

### Configure Nodes

Use ssh to configure ens7 on the nodes. Each node's dataplane connection will be assigned the IP address 192.168.<node_num>.100   


```python
node_num = 0
for node_name, node in experiment_topology.nodes.items():
    
    if node.name == router_name:
        print("skipping {}".format(node.name))
        continue
        
    management_ip = str(node.management_ip)
    #print("Node Name        : {}".format(node.name))
    #print("Management IP    : {}".format(management_ip))
    #print()
    
    
    
    nic_name = "nic" + str(node_num)
    net_name = "net" + str(node_num)
    router_iface_name = "iface" + str(node_num)
    node_num = node_num + 1
    node_dataplan_ip = "192.168." + str(node_num) + ".100"
    node_dataplan_cidr = '24'
    
    
    stdout = execute_script(username, node, 'sudo ip addr flush dev ens7')
    #print(stdout)
    
    print("{}, set ens7 to {}".format(node.name,node_dataplan_ip))
    stdout = execute_script(username, node, 'sudo ip addr add ' + node_dataplan_ip +'/' + node_dataplan_cidr + ' dev ens7')
    #print(stdout)
    
    stdout = execute_script(username, node, 'sudo ip link set dev ens7 up mtu 9000')
    #print(stdout)
```

### Configure Router

Use ssh to configure the ifaces on the router. This step requires testing the interfaces to figure out which interface is connected to which network.


```python
router_node = experiment_topology.nodes[router_name]

management_ip_router = str(router_node.management_ip)
print("Router Name        : {}".format(router_node.name))
print("Management IP    : {}".format(management_ip_router))
print()


iface_list = [ "ens7", "ens8", "ens9", "ens10" ]
node_num = 0
for node_name, node in experiment_topology.nodes.items():
    
    if node.name == router_name:
        print("skipping {}".format(node.name))
        continue
        
    management_ip = str(node.management_ip)
    #print("Node Name        : {}".format(node.name))
    #print("Management IP    : {}".format(management_ip))
    #print()
    
    nic_name = "nic" + str(node_num)
    net_name = "net" + str(node_num)
    router_iface_name = "iface" + str(node_num)
    node_num = node_num + 1
    node_dataplan_ip = "192.168." + str(node_num) + ".100"
    node_dataplan_cidr = '24'
    
    router_dataplan_ip = "192.168." + str(node_num) + ".1"
    router_dataplan_cidr = '24'
       
    
    for iface in iface_list:
        print("Testing, node_dataplan_ip: {}, iface: {}".format(node_dataplan_ip,iface),end='') 
        stdout = execute_script(username, router_node, 'sudo ip addr flush dev '+str(iface))
        #print(stdout)
    
        stdout = execute_script(username, router_node, 'sudo ip addr add ' + router_dataplan_ip +'/' + router_dataplan_cidr + ' dev '+str(iface))
        #print(stdout)
    
        stdout = execute_script(username, router_node, 'sudo ip link set dev '+ str(iface) + ' up mtu 9000')
        #print(stdout)
        
        if ping_test(router_node, node_dataplan_ip,verbose=False) > 0:
            #print(", success! keeping configuration")
            print(", success! Router {} -> {}".format(str(iface),net_name) )
            iface_list.remove(iface)
            #break
        else:
            print(", fail!")  #avg_rtt: {}".format(ping_test(router_node, node_dataplan_ip,verbose=False)))
        
        stdout = execute_script(username, router_node, 'sudo ip addr flush dev '+str(iface))
        #print(stdout)
    
```
